```python
import seaborn as sns
import pandas as pd
print ("All of dataset of seaborn: \n", sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.relplot(data = tips, x="total_bill", y = "tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset (using the 'tips' dataset)
tips = sns.load_dataset("tips")

# Create a relational plot (scatter plot) using relplot
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", style="time", size="size")

# Show the plot
plt.show()
```

```python
sns.set(font_scale = 1.2, style = 'darkgrid')
sns.scatterplot(x = 'total_bill', y ='tip', data = tips)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
tips = sns.load_dataset("tips")

# Create a scatter plot using scatterplot
sns.scatterplot(
    x="total_bill", y="tip", data=tips, hue="sex", style="time", size="size"
)

# Show the plot
plt.show()
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day')
plt.show()
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', hue = 'day', size = 'size')
plt.show()
```

```python
sns.relplot(data = tips, x='total_bill', kind = 'scatter', y='tip', hue = 'day', size = 'size', col = 'time')
plt.show()
```

```python
sns.relplot(data = tips, x='total_bill', y = 'tip', hue = 'day', size = 'size', kind = 'scatter', col = 'time', row = 'sex')
plt.show()
```
