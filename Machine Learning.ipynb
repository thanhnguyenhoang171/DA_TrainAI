```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
species = iris.target
x_reduced = PCA(n_components=3).fit_transform(iris.data)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with the reduced dimensions
ax.scatter(x_reduced[:, 0], x_reduced[:, 1], x_reduced[:, 2], c=species)

# Set labels and title
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("3D Scatter Plot of Iris Data (PCA)")

plt.show()
```

```python
import matplotlib.pyplot as plt
def Plot_faces(images, target,top_n):
    fig = plt.figure(figsize=(12, 12))
    for i in range(top_n):
        p=fig.add_subplot(20,20,i+1)
        p.imshow(images[i],cmap = 'gray')
        p.text(0,15,str(target[i]))
        
plt.show()

# Plot faces
Plot_faces(faces.images, faces.target, top_n=20)
```
